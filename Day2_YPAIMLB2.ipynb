```python
# Logical Operators
# and   or   not

# and operators requires atleast 2 conditions
# and will return TRUE only when both the conditions are TRUE

print(100==100 and 90<100)
#        c1          c2
#         T          T

print(100==100 and 90>100)
#        c1          c2
#         T          F

print(10==100 and 90>100)
#        c1          c2
#         F         F
```

```python
# or operators requires atleast 2 conditions
# or will return TRUE only when both the conditions are TRUE
# or will return TRUE even if any one of the condition is TRUE

print(100==100 or 90<100)
#        c1          c2
#         T          T

print(100==100 or 90>100)
#        c1          c2
#         T          F

print(10==100 or 90>100)
#        c1          c2
#         F         F
```

```python
# not used to inverse the result
print(not(100==100 or 90<100))
```

```python
# Membership Operator
# in   not in

# in return TRUE if the specified value exist in the Sequence

print("y" in "Python")
print("x" in "Python")
print("p" in "Python")

print(10 in [10,20,30,40])
```

```python
print("x" not in "Python")  # x does not exist in Python
print("y" not in "Python")  # y does not exist in Python
```

```python
# Find the largest
a = 19
b = 45
c = 34
d = 141
e = 69

largest = 0

if a>largest:      # 19>0 (t)
  largest = a      # largest = 19

if b>largest:     # 45>19 (t)
  largest = b     # largest = 45

if c>largest:    #  34>45(f)
  largest = c

if d>largest:   # 141>45 (t)
  largest = d   # largest = 141

if e>largest:
  largest = e

print('Largest Number is :',largest)
```

```python
# range()
# it is used to generate a sequence of integer values

# a.) range(x)
# it is used to generate a sequence of integer values
# staring from Zero till x-1

print(list(range(11)))

# b.) range(x,y)
# it is used to generate a sequence of integer values
# staring from x till y-1

print(list(range(1,11)))

# c.) range(x,y,stepCount)
# it is used to generate a sequence of integer values
# staring from x till y-1 based on stepCount

print(list(range(1,11,3)))

# print all the even numbers till 100 using this range()
print(list(range(2,101,2)))
```
