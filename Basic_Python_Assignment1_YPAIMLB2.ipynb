```python
# Python String capitalize()
# The capitalize() --> First character uppercase --> the rest lowercase

txt = "assalamualaikum, welcome to modern world"
a = txt.capitalize()
print (a)
```

```python
# Python String casefold()
# The casefold() --> all characters lowercase --> similar  lower()
# lower() > casefold() --> stronger convert more characters and more matches

txt = "Assalamualaikum, Welcome To Modern World"
b = txt.casefold()
print(b)
```

```python
# Python String count()
# The count() --> how many characters, single character and value in string

txt = "I enjoy learn python, python the best tool, python best language ever"
c = txt.count("python")
print(c)
```

```python
# Python String endswith()
# The endswith() --> true if string end with specific value
# --> while become false

txt = "Assalamualaikum, welcome to modern world"
d = txt.endswith("world")
print(d)
```

```python
# Python String find()
# The find() --> first occurrence of specified value
# --> return -1 if value not found --> almost same with index()
# --> difference is index() raises an exception if value is not found

txt = "Assalamualaikum, welcome to modern world"
e = txt.find("welcome")
print(e)
```

```python
# Python String islower()
# The islower() --> True if all characters lowercase, while false
# --> numbers,spaces, symbols not check only alphabet characters

txt = "modern world"
f = txt.islower()
print(f)
```

```python
# Python String isupper()
# The isupper() --> True if all characters uppercase, while false
# --> numbers,spaces, symbols not check only alphabet characters

txt = "FUTURE IS NOW"
g = txt.isupper()
print(g)
```

```python
# Python String lower()
# The lower() --> returns string where all characters lowercase
# --> numbers and symbols ignored

txt = "Assalamualaikum my FUTURE"
h = txt.lower()
print(h)
```

```python
# Python String upper()
# The upper() --> returns string where all characters uppercase
# --> numbers and symbols ignored


txt = "Assalamualaikum My Future"
i = txt.upper()
print(i)
```

```python
# Python String startswith()
# The startswith() --> true if string start with specific value
# --> while become false

txt = "Assalamualaikum, welcome to modern world"
j = txt.startswith("Assalamualaikum")
print(j)
```

```python
# Python String swapcase()
# The swapcase() --> returns a string where all uppercase letters
# --> are lowercase and vice versa

txt = "Morning My Name Is ASYRAF"
k = txt.swapcase()
print(k)
```

```python
# Python String replace()
# The replace() --> specified phrase with another specified phrase
# --> all occurrences of specified phrase will be replaced
# --> if nothing else is specified

txt = "I like python"
l = txt.replace("python", "programming")
print(l)
```
